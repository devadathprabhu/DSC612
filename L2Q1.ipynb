```python
import cv2
from skimage import io
from google.colab.patches import cv2_imshow
import numpy as np

img = io.imread("https://placekitten.com/800/571")
cv2_imshow(img)
```

```python
# rotate
height, width = img.shape[0:2]
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 0.5)
rotated_img = cv2.warpAffine(img, rotation_matrix, (width, height))
cv2_imshow(rotated_img)
```

```python
# resize
resized_img1 = cv2.resize(img, (0,0), fx=0.75, fy=0.75)
cv2_imshow(resized_img1)


resized_img2 = cv2.resize(img, (550, 350))
cv2_imshow(resized_img2)
```

```python
# crop
startRow = int(height*.15)
startCol = int(width*.15)
endRow = int(height*.85)
endCol = int(width*.85)
cropped_img = img[startRow:endRow, startCol:endCol]
cv2_imshow(cropped_img)
```

```python
# adjust the contrast
contrast_img = cv2.addWeighted(img, 2.5, np.zeros(img.shape, img.dtype), 0, 0)
cv2_imshow(contrast_img)
```

```python
# median blur
blur_img = cv2.medianBlur(img, 5)
cv2_imshow(blur_img)
```

```python
# conversion to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
# histogram equilization
src = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
histo_equi_img = cv2.equalizeHist(src)
cv2_imshow(histo_equi_img)
```

```python
# detect edges
edge_img = cv2.Canny(img, 100, 200)
cv2_imshow(edge_img)
```
